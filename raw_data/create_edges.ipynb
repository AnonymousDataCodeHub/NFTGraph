```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('./transactions.csv')
```

```python
df
```

```python
df.drop(columns=['token',],inplace=True)
```

```python
df['min'] = df[['source','target']].min(axis=1)
df['max'] = df[['source','target']].max(axis=1)
```

```python
df['Cnt'] = 0
```

```python
df
```

```python
df['Cnt'] = df.groupby(['min','max'])['Cnt'].transform("count")
df
```

```python
df['transferedAmount'] = df.groupby(['min','max'])['transferedAmount'].transform(sum)
df
```

```python
df['value'] = df.groupby(['min','max'])['value'].transform(sum)
df
```

```python
df['transactionFee'] = df.groupby(['min','max'])['transactionFee'].transform(sum)
df
```

```python
df
```

```python
# df = df.groupby(['min','max']).agg({'transferedAmount':sum,'value':sum,'transactionFee':sum,'Cnt':"count"}).reset_index(drop=False)
# df
```

```python
df.columns
```

```python
df = df.drop_duplicates(subset=['min','max'],ignore_index=True,keep='first')
```

```python
df['txn_id'] = pd.Series(np.arange(df.shape[0]))
```

```python
df.drop(columns=['min','max'],inplace=True)
```

```python
df.drop(columns=['txhash'],inplace=True)
```

```python
df.drop(columns=['edgelabel'],inplace=True)
```

```python
df.drop(columns=['tokenid'],inplace=True)
```

```python
df
```

```python
df = df.rename(columns={'Cnt':'TxnsCnt'})
```

```python
df
```

```python
dfout = df[['from','to','timestamp','transferedAmount','value','transactionFee','TxnsCnt']]
```

```python
dfout.to_csv('/data/sx/NFTGraph/edges.csv',index=False)
```
