```python
import numpy as np
import pandas as pd
```

```python
prefix = '/data/sx/NFTGraph'
```

```python
dfedges = pd.read_csv(prefix+'/raw_data/edges.csv')
```

```python
dfedges
```

```python
dfnodes = pd.read_csv(prefix+'/raw_data/nodes.csv')
```

```python
dfnodes
```

```python
dfnodes['Cnt'] = dfnodes['InCnt'] + dfnodes['OutCnt']
```

```python
dfnodes = dfnodes.sort_values('Cnt',ascending=False)
```

```python
topnodes = set(dfnodes.iloc[:20000,0].values.flatten())
```

```python
def seltopnode(line):
    return line['from'] in topnodes and line['to'] in topnodes
```

```python
tmp = dfedges.apply(seltopnode,axis=1)
```

```python
dfedges['topflag'] = tmp
```

```python
dfedges[dfedges['topflag']==1]
```

```python
dfedges = dfedges[dfedges['topflag']==True]
```

```python
dfedges.drop(columns=['topflag'],inplace=True)
dfedges
```

```python
dfnodes = dfnodes.iloc[:20000,:]
```

```python
dfnodes.drop(columns=['Cnt'],inplace=True)
```

```python
dfnodes.reset_index(drop=True,inplace=True)
```

```python
dfnodes
```

```python
dfnodes['label'].sum()
```

```python
dfnodes.to_csv(prefix+'/raw_data/tinynodes.csv',index=False)
```
